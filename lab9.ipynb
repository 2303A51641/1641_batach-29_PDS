```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Wine Quality dataset (red wine example)
# You can download it from UCI ML repo or use local file path
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')

# Select 5 key numerical features (commonly relevant chemical properties)
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol']

# Subset the dataframe
df_subset = df[features + ['quality']]

# Create pair plot with 'quality' as hue
sns.set(style="ticks")
pair_plot = sns.pairplot(df_subset, hue='quality', palette='viridis', diag_kind='kde', markers='o')

# Add title
plt.suptitle('Wine Quality Assessment: Pair Plot of Chemical Properties', y=1.02)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a custom dataset
data = {
    'study_time': [5, 3, 4, 2, 6, 7, 1, 4, 5, 3, 6, 2, 7, 4, 3],
    'attendance': [90, 80, 85, 75, 95, 100, 70, 85, 88, 82, 96, 78, 98, 87, 83],
    'exam_score': [88, 76, 85, 70, 92, 95, 68, 84, 89, 78, 93, 72, 97, 86, 79],
    'gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Optional: scale/normalize numerical data if needed (here, values are roughly on similar scales)

# Create pair plot with gender as hue
sns.set(style="ticks")
pair_plot = sns.pairplot(df, hue='gender', palette='Set2', diag_kind='kde', markers=['o', 's'])

# Add title
plt.suptitle('Student Performance: Study Time, Attendance, Exam Scores by Gender', y=1.02)

plt.show()
```
