```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('co2_emissions.csv')

# Check the first few rows (optional)
print(df.head())

# Plot CO2 Emissions over Year
plt.figure(figsize=(10,6))
plt.plot(df['Year'], df['CO2_Emissions'], marker='o', linestyle='-', color='green')

# Add title and axis labels
plt.title('Global CO2 Emissions Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('CO2 Emissions (Million Metric Tons)', fontsize=14)

# Add grid for readability
plt.grid(True, linestyle='--', alpha=0.5)

# Set x-axis limits and ticks for better scaling if necessary
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("student_scores.csv")

# Sort by Exam Number
data = data.sort_values("Exam_Number")

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(data["Exam_Number"], data["Score"], color='blue', marker='o', linewidth=2)

# Add title and axis labels
plt.title("Student Performance Over Exams", fontsize=16, fontweight='bold')
plt.xlabel("Exam Number", fontsize=12)
plt.ylabel("Score", fontsize=12)

# Grid and formatting
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(data["Exam_Number"])
plt.tight_layout()

# Show the chart
plt.show()
```
